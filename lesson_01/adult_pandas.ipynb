```python
import pandas as pd
```

```python
data = pd.read_csv("adult.data.csv")
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
data['sex'].value_counts()
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
print(data[data['sex'] == 'Female']['age'].mean())
print(data[data['sex'] == 'Female']['age'].median())

data[data['sex'] == 'Female']['age'].hist();
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
print((data['native-country'] == 'Germany').mean())
print(data[data['native-country'] == 'Germany'].shape[0] / data.shape[0])
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
less_than = data[data['salary'] == '<=50K']
more_than = data[data['salary'] == '>50K']

print('Для тех, кто получает >50K в год, mean = ', more_than['age'].mean(), ', std = ', more_than['age'].std())
print('Для тех, кто получает <=50K в год, mean = ', less_than['age'].mean(), ', std = ', less_than['age'].std())
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married_data = data[(data['marital-status'].str.startswith('Married')) & (data['sex'] == 'Male')]
not_married_data = data[(~data['marital-status'].str.startswith('Married')) & (data['sex'] == 'Male')]

print('Среди женатых мужчин', (married_data['salary'] == '>50K').mean())
print('Среди неженатых мужчин', (not_married_data['salary'] == '>50K').mean())
#Доля зарабатывающих >50K среди женатых мужчин больше
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data['hours-per-week'].max()
print('Максимальное число часов в неделю:', max_hours)

max_hours_data = data[data['hours-per-week'] == max_hours]
print('В неделю столько часов работает', max_hours_data.shape[0], 'человек')

print('Процент зарабатывающих много', round((max_hours_data['salary'] == '>50K').mean() * 100, 2), '%')
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
g = data.groupby(['native-country', 'salary'])

for (i, sub_df) in g:
    print(i[0], i[1], round(sub_df['hours-per-week'].mean(), 2))
```
